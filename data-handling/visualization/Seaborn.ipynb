## Seaborn

- 기존 matplotlib에 기본 설정을 추가
- 복잡한 그래프를 간단하게 만들 수 있는 wrapper
- 간단한 코드 + 예븐 결과
- 다양한 그래프를 한 화면에 쉽게 보여줌

## Seaborn - basic plots

- matplotlib와 같은 기본적은 Plot
- 손쉬운 설정으로 데이터 산출
- lineplot, scatterplot, countplot 등

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset('tips')
fmri = sns.load_dataset('fmri')
```

```python
tips
```

```python
fmri
```

```python
# Plot the responses for different events and regions
sns.set_style('whitegrid')
sns.lineplot(x='timepoint', y = 'signal', data = fmri)
```

```python
sns.set_style('dark')
sns.lineplot(x='timepoint', y='signal', data=fmri)
```

```python
fmri.sample(n=10, random_state=1)
```

```python
# hue를 넣으면 해당하는 값들의 평균을 따로 구해줄 수 있다
sns.lineplot(x='timepoint', y='signal', hue='event', data=fmri)
```

## seaborn - basic plots

### scatterplot

```python
# hue를 넣지 않았을 때
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# hue를 넣었을 때 - 색깔별로 나오게 됨
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

### countplot

- 단순히 데이터의 개수만 세어줌
- 주로 카테고리 데이터의 개수를 셈

```python
tips
```

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.countplot(x='smoker', hue='time', data=tips)
```

```python
fmri
```

```python
sns.countplot(data=fmri, x='event', hue='region')
```
