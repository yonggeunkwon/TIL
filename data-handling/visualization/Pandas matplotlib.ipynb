## Pandas matplotlib

- matplotlib를 사용한 그래프 지원
- Dataframe, series별로 그래프 작성 가능

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_url = './data/housing.data'
df_data = pd.read_csv(data_url, sep='\s+', header = None) # csv 타입 데이터 로드, separate는 빈공간으로 지정하고 Column은 없음
df_data.columns = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', "NOX", 'RM', 'AGE', "DIS", 'RAD', "TAX", "PTRATIO", 'B', 'LSTAT', 'MEDV']
# Column Header 이름 지정
df_data.head()
```

```python
fig = plt.figure()
ax = []
for i in range(1, 5):
    ax.append(fig.add_subplot(2, 2, i))
    
ax[0].scatter(df_data["CRIM"], df_data["MEDV"], s=10)
ax[1].scatter(df_data["PTRATIO"], df_data["MEDV"], s=10)
ax[2].scatter(df_data["AGE"], df_data["MEDV"], s=10)
ax[3].scatter(df_data["NOX"], df_data["MEDV"], s=10)

plt.show()
```

```python
fig = plt.figure()
ax = []
for i in range(1, 5):
    ax.append(fig.add_subplot(2, 2, i))
    
columns = ["CRIM", "PTRATIO", "AGE", "NOX"]
colors = ["b", "g", "c", "r"]
for i in range(4):
    ax[i].scatter(
        df_data[columns[i]], df_data["MEDV"],s=0.3, color = colors[i], label = columns[i])
    ax[i].legend()
    ax[i].set_title(columns[i])
    
plt.subplots_adjust(wspace = 0, hspace = 0)
plt.show()
```

## pandas matplotlib - plot function

- plot 함수를 사용하면 전체 데이터의 graph를 그림
- kind : plot whdfb
- x : x축 컬럼
- y ; y축 컬럼

```python
# plot은 다 보여줌
df_data.plot()
plt.show()
```

```python
# series 데이터로 바꿔서 부르기 (조금 더 일반적인 방법)
df_data["AGE"].plot.hist(bins=12, alpha=0.5)
plt.show()
```

```python
df_data["AGE"].plot(kind='line')
plt.show()
```

```python
df_data["AGE"].sort_values().reset_index(drop=True).plot(kind='line')
```

```python
# age column을 sorting하고, 기존의 인덱스를 없애줌
df_data["AGE"].sort_values().reset_index(drop=True)
```

```python
df_data.plot(kind='scatter', x = 'AGE', y = 'MEDV')
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scale_data = std_scaler.fit_transform(df_data)
scale_data
```

```python
corr_data = np.corrcoef(scale_data.T[:5])
corr_data
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_data, vmin = -1, vmax = 1, interpolation = 'nearest')
fig.colorbar(cax)
fig.set_size_inches(10, 10)
ticks = np.arange(0, 5, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(df_data.columns[:5])
ax.set_yticklabels(df_data.columns[:5])

plt.show()
```
