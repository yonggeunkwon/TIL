## Scatter

- scatter 함수 사용, marker: scatter 모양 지정
- numeric한 데이터인 경우, 둘의 상관관계를 보기 위해서 사용
- 지도 데이터 활용 가능

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Random한 두 개의 데이터 생성
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)
```

```python
data_1
```

```python
# x축과 Y축의 데이터를 넣고, 색깔과 마커의 모양을 넣으면 점을 찍어줌
plt.scatter(data_1[:,0], data_1[:, 1], c="b", marker = 'x')
plt.scatter(data_2[:,0], data_2[:, 1], c="r", marker = '^')

plt.show()
```

```python
N = 50
a = np.random.rand(N)
b = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2
```

```python
a.shape, b.shape, colors.shape
```

```python
# 지도 데이터와 합칠 수 있음
plt.scatter(a, b, c = colors, alpha=0.5)
plt.show()
```

## Bar chart

- Bar 함수 사용

```python
data = [[5., 25., 50., 20.],
       [4., 23., 51., 17.],
       [6., 22., 52., 19.]]
```

```python
# width는 bar의 너비
X = np.arange(4)
plt.bar(X + 0.00, data[0], color = 'b', width = 0.15)
plt.bar(X + 0.25, data[1], color = 'g', width = 0.15)
plt.bar(X + 0.50, data[2], color = 'r', width = 0.15)
plt.xticks(X+0.25, ("A", "B", "C", "D"))

plt.show()
```

### 위로 쌓는 Bar chart

```python
data = [[5., 25., 50., 20.],
       [4., 23., 51., 17.],
       [6., 22., 52., 19.]]
```

```python
# axis = 0 이므로 위로 쌓임
color_list = ['b', 'g', 'r']
data_label = ["A", "B", "C"]
for i in range(3):
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis = 0),
           color = color_list[i], label=data_label[i])
plt.legend()
plt.show()
```

```python
A = [5., 30., 45., 22.]
B = [5, 25, 50, 20]

X = range(4)

plt.bar(X, A, color = 'b')
plt.bar(X, B, color = 'r', bottom = 60)
plt.yticks([0, 60], ['A', 'B'])
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')

plt.show()
```
