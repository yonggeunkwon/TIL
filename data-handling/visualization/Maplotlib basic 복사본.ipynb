### Matplotlib

- 최대 단점 argument를 kwargs 받음
- 고정된 argument가 없어서 alt+tab으로 확인이 어려움
- Graph는 원래 figure 객체에 생성됨 (figure는 도화지라고 생각하면 됨)
- pyplot 객체 사용시 기본 figure에 그래프가 그려짐

```python
# cos graph
import numpy as np
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.plot(range(100), range(100))
plt.show() # plt.show()를 하면, 메모리에서 없애면서 그림을 그려줌
```

```python
fig = plt.figure() #figure 반환
fig.set_size_inches(10, 5) # figure의 크기 지정. 크기에 따라 그래프의 크기가 바뀜
ax_1 = fig.add_subplot(1, 2, 1) # 두 개의 Plot 생성 (첫번째 칸에 두개의 줄이 있는데 그 중에 첫 번쨰)
ax_2 = fig.add_subplot(1, 2, 2) # 두 개의 plot 생성 (첫번째 칸에 두개의 줄이 있는데 그 중에 두 번쨰)

ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
plt.show()
```

### subplots

- Subplot의 순서를 grid로 지정

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10, 10) # figure 크기 지정

ax_1 = fig.add_subplot(2, 2, 1) # 두 칸 그리고, 첫번째 칸에서 천번째 줄
ax_2 = fig.add_subplot(2, 2, 2) # 두 칸 그리고, 첫번째 칸에서 두번째 줄
ax_3 = fig.add_subplot(2, 1, 2) # 두 칸 그리고, 두번째 칸 다 쓸 거임

ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
ax_3.plot(X_1, Y_1)
plt.show()
```

```python
# loop을 통해 여러 개의 도화지 handling

import numpy as np
fig = plt.figure() # figure 반환
fig.set_size_inches(10, 10) # figure 크기 지정
plt.style.use("ggplot") # 스타일 적용

ax = []
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i in range(1, 7):
    ax.append(fig.add_subplot(2, 3, i)) # 두개의 Plot 생성
    X_1 = np.arange(50)
    Y_1 = np.random.rand(50)
    c = colors[np.random.randint(1, len(colors))]
    
    ax[i-1].plot(X_1, Y_1, c=c)
```

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10, 10) # 크기 지정

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 1, 2)

ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
ax_3.plot(X_1, Y_1)

ax_2.margins(2, 5) # 두번째 plot에 margin 주기

left = 0.125
right = 0.9
bottom = 0.1
top = 0.9
wspace = 0.5
hspace = 0.1

fig.subplots_adjust(left, bottom, right, top, wspace, hspace=None)

plt.show()
```

### 월별 통화량 데이터 총합 구해보기

```python
import pandas as pd
df = pd.read_csv("./data/phone_data.csv")
df.head()
```

```python
# 월별 통화량(duration)의 합
result = df.groupby(["month"])["duration"].sum()
result
```

```python
# x축에는 result의 index가, y축에는 result (series 데이터)가, 색깔과 linestyle 지정
plt.plot(result.index, result, color = 'b', linestyle="dashed")
plt.title("Duration per month")
plt.show()
```

## Set color

## Set legend

- Legend 함수로 범례를 표시함, loc 위치 등 속성 지정

```python
plt.plot(X_1, Y_1, color = 'b', linestyle = "dashed", label = 'line_1')
plt.plot(X_2, Y_2, color = 'r', linestyle = "dotted", label = 'line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")
```

- xlabel, ylabel 값을 넣을 수 있음

```python
plt.plot(X_1, Y_1, color = 'b', linestyle = "dashed", label = 'line_1')
plt.plot(X_2, Y_2, color = 'r', linestyle = "dotted", label = 'line_2')

plt.title('$y = ax + b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')

plt.show()
```

```python
plt.plot(X_1, Y_1, color = 'b', linestyle = "dashed", label = 'line_1')
plt.plot(X_2, Y_2, color = 'r', linestyle = "dotted", label = 'line_2')

plt.annotate(
    'line_2', xy =(50, 150), xytext=(20, 175),
    arrowprops=dict(facecolor = 'black', shrink = 0.05))
plt.annotate(
    'line_1', xy =(50, 50), xytext=(20, 75),
    arrowprops=dict(facecolor = 'red', shrink = 0.05))


plt.title('$y = ax + b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')

plt.show()
```

## Set grid & xylim

- Graph 보조선을 긋는 Grid와 xy축 범위 한계를 지정

```python
plt.plot(X_1, Y_1, color = 'b', linestyle = "dashed", label = 'line_1')
plt.plot(X_2, Y_2, color = 'r', linestyle = "dotted", label = 'line_2')
plt.legend(shadow = True, fancybox = True, loc = "lower right")

plt.grid(True, lw = 0.7, ls = "-", c = ".10")
# plt.xlim(-1000, 2000)
# plt.ylim(-1000, 2000)

plt.show()
```

```python
plt.plot(X_1, Y_1, color = 'b', linestyle = "dashed", label = 'line_1')
plt.plot(X_2, Y_2, color = 'r', linestyle = "dotted", label = 'line_2')

plt.grid(True, lw = 0.7, ls = "-", c = ".10")
plt.legend(shadow = True, fancybox = True, loc = "lower right")
plt.xlim(-1000, 2000)
plt.ylim(-1000, 2000)
plt.savefig("test.png")

plt.show()
```

## 일별 네트워크별 데이터의 총량을 다중 Plot으로 작성하라

```python
def parsedate(date_info):
    import datetime
    return date_info[:9]
```

```python
df["day"] = df["date"].map(parsedate)
df.head()
```

```python
result = df.groupby(["network", "day"])["duration"].sum().reset_index()
result.head()
```

```python
result["network"].unique().tolist()
```

```python
result[result["network"] == 'data'].head()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10) # 사이즈 설정
# plt.style.use("ggplot") # 스타일 적용

network_types = result["network"].unique().tolist()
ax= []
for i in range(1,7):
    ax.append(fig.add_subplot(2,3,i)) # 두개의 plot 생성
    network_name = network_types[i-1]
    
    plt.title(network_name)
    
    X_1 = result[result["network"] == network_name]["day"]
    Y_1 = result[result["network"] == network_name]["duration"]
    ax[i-1].get_xaxis().set_visible(False)
    ax[i-1].plot(X_1, Y_1)
```
