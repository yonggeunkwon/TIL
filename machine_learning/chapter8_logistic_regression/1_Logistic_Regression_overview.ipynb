## Classification Problem

- 회귀든 분류든, 둘 다 선을 긋는 문제임
- 분류 문제에서는 의사결정선을 어떻게 찾을 것인지가 classification의 핵심 문제

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./data/generator.csv')
df.head()
```

```python
fig, ax = plt.subplots()

color = ['blue', 'red']
status = ['good', 'faulty']
for i in range(2):
    x = df[df['STATUS'] == status[i]]['RPM']
    y = df[df['STATUS'] == status[i]]['VIBRATION']
    ax.scatter(x, y, c=color[i], s=50, label=status[i], alpha=0.3, edgecolors='none')

x_line = df['RPM']
y_line = 830 - 0.667 * df['RPM']
ax.plot(x_line, y_line, linestyle='-', color='green', linewidth=1, label='line')

ax.legend()
plt.title('Vibration = 830 - 0.667 * RPM')
plt.show()
```

#### Regression 문제와 똑같은 방법으로 학습하는 방법

$f_(x) = 830 - 0.667 * RPM - 1 * Vibration $ 
- 0.667과 -1은 각각 weight 값
- $f_(x) >= 0$ 이면 1
- 이외에는 0

```python
x_data = df.iloc[:, 1:3].to_numpy()

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y_label = df.iloc[:, 3].to_numpy()
le.fit(np.unique(y_label))
y_data = le.transform(y_label)

x_data[:3], y_data[:3]
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(x_data, y_data)
```

```python
regr.predict(x_data)
```

```python
fig, ax = plt.subplots()

color = ["blue", "red"]
status = ["good", "faulty"]
for i in range(2):
    x = df[df["STATUS"] == status[i]]["RPM"]
    y = df[df["STATUS"] == status[i]]["VIBRATION"]
    ax.scatter(x, y, c=color[i], s=50, label=status[i], alpha=0.3, edgecolors='none')


x_line = x_data[:, 0]
y_line = -1 * (regr.intercept_ + regr.coef_[0] *  x_data[:,0]) / regr.coef_[1]
ax.plot(x_line, y_line, color="green", linewidth=1, label="line")


ax.legend()

plt.show()
```

#### 기존의 Regression 방법을 썼을 때 발생하는 문제점들

- 0과 1사이의 범위를 벗어나는 수들을 어떻게 해석할 것인가?
- 1 또는 0으로 정확히 표현이 가능한가?
- 변수가 y에 영향을 주는 정도가 비례하는가?
- 따라서 확률로 발생할 사건의 가능성을 표현해야 함
