## Gradient descent based learning

- 실제 값과 학습된 모델 예측치의 오차를 최소화
- 모델의 최적 parameter 찾기가 목적

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 1)
f_x = x ** 2

plt.plot(x, f_x)
plt.show()
```

```python
x_new = 10
derivative = []
y = []
learning_rate = 0.1
for i in range(100):
    old_value = x_new
    derivative.append(old_value - learning_rate * 2 * old_value)
    x_new = old_value - learning_rate * 2 * old_value
    y.append(x_new ** 2)
```

```python
plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```

```python
def sin_function(x):
    return x * np.sin(x ** 2) + 1

def derivative_f(x):
    return np.sin(x ** 2) + 2 * (x ** 2) * np.cos(x ** 2)
```

```python
x = np.arange(-3, 3, 0.001)
f_x = sin_function(x)
```

```python
plt.plot(x, f_x)
plt.show()
```
