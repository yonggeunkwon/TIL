## Gradient descent(경사하강법)

- Xnew = Xold - a x (Xold미분)
- 결국은 미분인데, 컴퓨터가 쉽게 이해할 수 있도록 한 것

### 정해야 하는 것

- Learning rate (알파)에 대한 선정
- 얼마나 많이 loop을 돌지에 대한 선정

#### Learning Rate나 loop의 수가 너무 작은 경우
- 시간이 너무 오래 걸리고, 원하는 지점까지 도달하지 못할 수 있음

#### Learning Rate은 작고, loop의 수가 너무 큰 경우
- 시간이 아주 오래 걸림

#### Learning Rate가 너무 큰 경우
- 오히려 값이 수렴하지 않고 발산할 수 있음

### 굴곡이 많은 함수에서의 Gradient descent
- 시작점에 따라서 최적점을 찾지 못하는 경우가 발생할 수 있다.
- 따라서 시작점의 선택이 중요하다.
- 단순 Gradient descent로는 위의 문제를 해결할 수 없다. (stochastic 등의 방법을 활용해야함)

### Gradient descent example

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data 생성
x_data = np.array([8759, 10132, 12078, 16430])
y_data = np.array([487, 612, 866, 1030])

# 그래프로 나타내기
plt.scatter(x_data, y_data, color='r', marker='s')
plt.show()
```

```python
# 임의의 parameter 생성
weight_0 = np.random.sample(5)
weight_1 = np.random.sample(5)

weight_0, weight_1
```

```python
# y_hat = w0 + w1 * x_data
y_hat_list = []
for w0, w1 in zip(weight_0, weight_1):
    y_hat_list.append(w0 + w1 * x_data)

y_hat_list
```

```python
plt.scatter(x_data, y_data, color='r', marker='s')
for y_hat in y_hat_list:
    plt.plot(x_data, y_hat)

plt.show()
```

### Linear regression with Gradient descent

- 임의의 세타0, 세타1 값으로 초기화
- Cost Function이 최소화 될 때까지 학습
- 더 이상 cost function이 줄어들지 않거나 학습 횟수를 초과할 때 종료

- Learning rate, Iteration 횟수 등 Parameter를 사람이 직접 지정해줘야 하는 단점이 있음
- Feature가 많으면 Normal equation에 비해 상대적으로 빠름
- 최적값에 수렴하지 않을 수도 있음
