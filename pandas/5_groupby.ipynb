### Group By

- split, apply, combine

```python
import pandas as pd
```

### Group by - Basic

```python
# data from: 
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'Kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
# Team을 기준으로(Team별로) Points의 값을 sum 해달라
df.groupby("Team")["Points"].sum()
```

```python
df.groupby("Team")["Points"].mean()
```

### Groupby 2

- Groupby에 의해 Split된 상태를 추출 가능함

```python
# key, value 형태로 뽑을 수 있음
grouped = df.groupby("Team")

for name, group in grouped:
    print(name)
    print(group)
```

```python
list(grouped)
```

```python
# grouped 상태에서 특정 key 값을 가진 그룹의 정보만 추출하기
grouped.get_group("Devils")
```

### grouped 된 상태에서 할 수 있는 것들 : 추출된 group 정보에는 세 가지 유형의 Apply가 가능함

- Aggregation: 요약된 통계정보를 추출해줌
- Transformation: 해당 정보를 변환해줌
- Filtration: 특정 정보를 제거하여 보여주는 필터링 기능


### Groupby - aggregation

```python
# agg - aggregation의 약자
grouped.agg(sum)
```

```python
import numpy as np
grouped.agg(np.mean)
```

```python
# 특정 column에 여러개의 fucntion을 apply 할 수도 있음
grouped['Points'].agg([np.sum, np.mean, np.std])
```

```python
# grouped에서 각 column별로 가장 큰 값들이 나옴
grouped.agg(max)
```

### Groupby - transformation

- Aggregation과 달리 key값 별로 요약된 정보가 아님
- 개별 데이터의 변환을 지원함
- data 변환시킬 때 많이 사용하게 됨

```python
df
```

```python
# transformation은 group된 상태에서 모든 값들을 변환함
# ex) (위의 df 참고) 0, 1, 8, 11번 인덱스는 Riders를 기준으로 그룹이므로, 그 둘 중에서 column 중 큰 값을 모두 적용시킴
# 즉, Rank에서는 1, 2 중에서 2가 크므로 모두 다 2가 적용되고, Year에서는 2017이 가장 크므로 모두 2017이 적용됨
score = lambda x: (x.max())
grouped.transform(score)
```

```python
# score = 정규화 시켜주는 식
score = lambda x: (x - x.mean()) / x.std()
grouped.transform(score)
```

```python
# min, max 정규화
# 최대값을 1, 최소값을 0으로 반환시켜주는 식
score = lambda x: (x - x.min()) / (x.max()- x.min())
grouped.transform(score)
```

### Groupby - filter

- 특정 조건으로 데이터를 검색할 때 사용함
- filter 안에는 Boolean 조건이 존재해야함
- len(x)는 grouped 된 DataFrame 개수

```python
df
```

```python
# team이 3개 이상의 데이터가 존재하는 팀만 뽑기
df.groupby('Team').filter(lambda x: len(x) >= 3)
```

```python
df["Team"].value_counts()
```

```python
# 각 팀별로 Points의 합이 2000 이상인 것들만 출력하기
df.groupby('Team').filter(lambda x: x["Points"].sum() > 2000)
```

```python
# 각 팀별로 평균이 750 이상인 팀들만 출력
df.groupby("Team").filter(lambda x: x["Points"].mean() > 750)
```
